```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
#to display all columns in the dataset
pd.set_option('display.max_columns',None)
df=pd.read_csv('E:/my_work_my_work/data analysis project/prosperLoanData/prosperLoanData.csv')
df.head()
```

## (Visualization 1)

bar chart of LoanStatus.

Observations:

1-number of completed loan=38074.

2-number of loans are in current pending state=56576.

3-number of loans have past due(1 to 15 days)=806.

4-number of loans chargedoff=11992.

5-number of loans are defaulted=5018.


```python
plt.figure(figsize=(15,10))
sb.countplot(x='LoanStatus',data=df)
plt.xticks(rotation = 90)
```

## (Visualization 2)

ProsperRating (Alpha) vs IsBorrowerHomeowner:-

  Observations:-

           Home improvement for all even if you do not have home.

```python
sb.countplot(data=df,x='ProsperRating (Alpha)',hue='IsBorrowerHomeowner',palette='Set2')
plt.legend(loc='lower left',bbox_to_anchor=(1.0,0.5))
plt.legend(['no_home','has_home'])
```

## (Visualization 3)
LoanStatus Vs BorrowerAPR VS EmploymentStatus.

Observations:

 The employees receive loans fewer than the unemployed.


```python
plt.figure(figsize=[23,15])
sb.violinplot(x="LoanStatus", y="BorrowerAPR", hue='EmploymentStatus', data=df);
plt.xticks(rotation = 90);
```
